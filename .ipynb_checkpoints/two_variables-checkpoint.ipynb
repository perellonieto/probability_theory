%matplotlib inline

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 16, 12
```

Weight vs Height
================

Artificially created data simulating weight and height in a reduced interval

```python
N = 100
gaussian = np.random.normal(0,1,(N,2))
covariance = np.array([[1, 2], [3, 2]]).T
center = np.array([60,170])
points = np.dot(gaussian,covariance) + center
X = weight = points[:,0]
Y = height = points[:,1]
```

Marginal Probabilities
======================

```python
plt.subplot(2, 2, 1)
plt.plot(weight, height,'b.')
plt.xlabel('weight')
plt.ylabel('height')

plt.subplot(2, 2, 2)
plt.hist(height, normed=True, orientation=u'horizontal')
plt.title('P(Height)')
plt.ylabel('height')
plt.subplot(2, 2, 3)
plt.hist(weight, normed=True)
plt.title('P(Weight)')
plt.xlabel('weight')
```

Conditional Probabilities
=========================

```python
def marginal(X, Y, y_min, y_max):
    return [X[i] for i,y in enumerate(Y) if y >= y_min and y <= y_max]
```

```python
plt.subplot(1, 2, 1)
plt.plot(weight, height,'b.')
plt.xlabel('weight')
plt.ylabel('height')

x_y_155_160 = marginal(X,Y,155,160)
x_y_160_165 = marginal(X,Y,160,165)
x_y_165_170 = marginal(X,Y,165,170)
x_y_170_175 = marginal(X,Y,170,175)
x_y_175_180 = marginal(X,Y,175,180)

plt.subplot(5, 2, 10)
plt.hist(x_y_155_160, normed=True)
plt.ylabel('P(X|y in [155,160])')
plt.xlim(50,68)
plt.subplot(5, 2, 8)
plt.ylabel('P(X|y in [160,165])')
plt.xlim(50,68)
plt.hist(x_y_160_165, normed=True)
plt.subplot(5, 2, 6)
plt.ylabel('P(X|y in [165,170])')
plt.xlim(50,68)
plt.hist(x_y_165_170, normed=True)
plt.subplot(5, 2, 4)
plt.ylabel('P(X|y in [170,175])')
plt.xlim(50,68)
plt.hist(x_y_170_175, normed=True)
plt.subplot(5, 2, 2)
plt.ylabel('P(X|y in [175,180])')
plt.xlim(50,68)
plt.hist(x_y_175_180, normed=True)
```

Probability distribution
========================

Differences in the probability distribution when the discretization of the space changes. As we decrease the size of the bins we see that some of the probabilities starts bannishing. In the extreme with the continuous space discretized some of the probabilities becomes zero, altough it is in the middle of the Gaussian shape.

```python
BINS = np.multiply(range(1,11),15)
for i, bins in enumerate(BINS):
    plt.subplot(10,2,(i*2)+1)
    plt.hist(weight, normed=True, bins=bins)
    
    y,binEdges=np.histogram(weight,bins=bins)
    bincenters = 0.5*(binEdges[1:]+binEdges[:-1])
    #from scipy.interpolate import interp1d
    #f2 = interp1d(y, bincenters, kind='cubic')
    plt.subplot(10,2,(i*2)+2)
    plt.plot(bincenters,y,'-')
```
